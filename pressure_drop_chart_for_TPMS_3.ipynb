```python
import matplotlib.pyplot as plt
import numpy as np

# Data
tpms_structures = ['Diamond', 'Gyroid', 'Primitive', 'G-Prime 1', 'FRD', 'G-Prime 2', 'FRD Prime']
pressure_drops = [4.9, 9.18, 5.97, 12.18, 20.27, 25.11, 26.27]
nusselt_numbers = [124.73, 129.87, 0, 165.52, 182.25, 212.26, 247.96]

# Create figure and axes
fig, ax = plt.subplots(figsize=(12, 10))

# Scatter plot with structure names
scatter = ax.scatter(pressure_drops, nusselt_numbers, c=np.arange(len(tpms_structures)), cmap='viridis', s=200)
for i, tpms in enumerate(tpms_structures):
    ax.annotate(tpms, (pressure_drops[i], nusselt_numbers[i]), xytext=(10, 10), textcoords="offset points", fontsize=12)

# Add gridlines, labels, and title
ax.grid(True, linestyle='--', linewidth=1)
ax.set_xlabel('Pressure Drop (Pa)', fontsize=14)
ax.set_ylabel('Nusselt Number (Nu)', fontsize=14)
ax.set_title('Comparative Performance of TPMS Structures', fontsize=16)

# Add colorbar
cbar = fig.colorbar(scatter, ticks=np.arange(len(tpms_structures)))
cbar.ax.set_yticklabels(tpms_structures)
cbar.ax.set_title('TPMS Structure', fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
tpms_structures = ['Diamond', 'Gyroid', 'Primitive', 'G-Prime 1', 'FRD', 'G-Prime 2', 'FRD Prime']
pressure_drop_coefficients = [8.003989866187, 14.991382919, 9.746525283, 21.4550125117, 38.67199487523, 40.9949807182, 42.88903756731]
nusselt_numbers = [124.73, 129.87, 0, 165.52, 182.25, 212.26, 247.96]

# Create figure and axes
fig, ax = plt.subplots(figsize=(12, 10))

# Plot quadrant divisions
ax.axvline(x=20, color='gray', linestyle='--')
ax.axhline(y=150, color='gray', linestyle='--')

# Color the quadrants
ax.fill_between(np.linspace(0, 20, 100), 150, 300, color='green', alpha=0.2)
ax.fill_between(np.linspace(20, 50, 100), 150, 300, color='yellow', alpha=0.2)
ax.fill_between(np.linspace(0, 20, 100), 0, 150, color='orange', alpha=0.2)
ax.fill_between(np.linspace(20, 50, 100), 0, 150, color='red', alpha=0.2)

# Scatter plot with structure names
scatter = ax.scatter(pressure_drop_coefficients, nusselt_numbers, c=np.arange(len(tpms_structures)), cmap='viridis', s=200)

for i, tpms in enumerate(tpms_structures):
    ax.annotate(tpms, (pressure_drop_coefficients[i], nusselt_numbers[i]), xytext=(10, 10), textcoords="offset points", fontsize=12)

# Add gridlines, labels, and title
ax.grid(True, linestyle='--', linewidth=1)
ax.set_xlabel('Pressure Drop Coefficient', fontsize=14)
ax.set_ylabel('Nusselt Number (Nu)', fontsize=14)
ax.set_title('Comparative Performance of TPMS Structures', fontsize=16)

# Add colorbar at the top
cbar_ax = fig.add_axes([0.15, 0.92, 0.7, 0.03])
cbar = fig.colorbar(scatter, cax=cbar_ax, orientation='horizontal', ticks=np.arange(len(tpms_structures)))
cbar.ax.set_xticklabels(tpms_structures)
cbar.ax.set_title('TPMS Structure', fontsize=12)

plt.show()
```
